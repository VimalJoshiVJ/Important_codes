```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
A= pd.read_excel('BN-Coronene B3LYP.xlsx' , engine= 'openpyxl')
B= pd.read_excel('BN-Coronene wB97X.xlsx' , engine= 'openpyxl')
```

```python
A
```

```python
B
```

```python
plt.vlines(A['X'], ymin=0 , ymax=A['Y'] , color= 'blue', label= 'BN-Coronene B3LYP-D3BJ')
plt.vlines(B['X'] ,ymin=0, ymax=B['Y'], color= 'black', label= 'BN-Coronene wB97X-D3BJ')
plt.xlim(0,4000)
plt.ylim(110,0)
plt.yticks(fontweight='bold')
plt.xlabel("Frequency (cm$^{-1}$)", fontweight='bold')
plt.ylabel("Intensity", fontweight='bold')
plt.legend( loc="lower right")
ax = plt.gca()
ax.xaxis.tick_top()                  # move x-axis ticks to the top
plt.xticks(fontweight='bold')
ax.xaxis.set_label_position('top')   # move x-axis label to the top 
for spine in plt.gca().spines.values():
    spine.set_linewidth(2.5)
    spine.set_color('black') 
plt.savefig("BN-Coronene Freq Plot")
```
